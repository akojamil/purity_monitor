```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ring = np.arange(1,10,1)
voltage = [0.234, 0.270, 0.303, 0.333, 0.363, 0.397, 0.436, 0.488, 0.673]
voltage_calc = [0.281,0.334,0.379,0.428,0.487,0.562,0.660,0.797,0.293]
```

```python
figure = plt.figure(figsize=(8,6))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(-0.8, 0.8)
plt.xlabel('Distance from Cathode Grid [cm]', fontsize=16)
plt.ylabel('Voltage Drop [V]', fontsize=16)
plt.scatter(ring, voltage, label='Measured')
plt.scatter(np.arange(1,10,1), voltage_calc, label='Calculated')
plt.scatter(-1, -0.689)
plt.legend(loc='lower right', fontsize=16)
plt.title('Cathode and Anode Biased to 20V', fontsize=16)
```
